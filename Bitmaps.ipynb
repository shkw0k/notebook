```python
import PIL.Image as Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
def readJpegRGB (fname):
    global r,g,b
    img = Image.open(fname)
    cols,rows = img.size
    r,g,b = img.split()
    red = np.array(r.getdata()).reshape((rows,cols))
    grn = np.array(g.getdata()).reshape((rows,cols))
    blu = np.array(b.getdata()).reshape((rows,cols))
    return red, grn, blu
```

```python
fileName = "../images/smarties-loose.jpg"
```

```python
red, grn, blu = readJpegRGB(fileName)
```

```python
plt.figure(figsize=(14,6))
plt.subplot(131)
plt.imshow(red, cmap='gray')
plt.subplot(132)
plt.imshow(grn, cmap='gray')
plt.subplot(133)
plt.imshow(blu, cmap='gray')
```
