```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import imresize
import struct, csv, mmap, time
```

```python
DataDir = "../bigisland/"
```

```python
class BILReader:
    HDREXT = ".hdr"
    BILEXT = ".bil"
    
    def __init__(self, baseName):
        self.baseName = baseName
        self.data = self.readHeader ()
    
    def readHeader(self):
        fname = self.baseName + self.HDREXT
        with open(fname, "r") as fh:
            rawHeaders = csv.reader(fh, delimiter=" ")
            headers = {}
            for h in rawHeaders:
                h1 = [ x for x in h if x != "" ]
                if len(h1) == 2:
                    headers[h1[0]] = h1[1]
            self.headers = headers
            self.sanityCheck()
    
    def sanityCheck (self):
        headers = self.headers
        if headers["BYTEORDER"] != "I":
            raise Exception("Byte order must be I")
        if float(headers["NROWS"]) < 0:
            raise Exception("NROWS must be greater than 0")
        if float(headers["NCOLS"]) < 0:
            raise Exception("NCOLS must be greater than 0")
        if headers["LAYOUT"] != "BIL":
            raise Exception("LAYOUT must be BIL")
    
    def getHeaders (self):
        return self.headers
    
    def readSlow (self):
        fname = self.baseName + self.BILEXT
        width = int(self.headers["NCOLS"])
        height = int(self.headers["NROWS"])
        nbytes = int(self.headers["NBITS"]) / 8
        
        with open(fname, "rb") as fh:
            if fh:
                print (time.clock())
                mfh = mmap.mmap(fh.fileno(), 0, access=mmap.ACCESS_READ)
                cont=fh.read()
                print ("read done", time.clock())
                desc = "<%dH" % (int(width*height),)
                data = struct.unpack (desc, cont)
                print ("convert done", time.clock())
                return data
            else:
                return None

    def read (self):
        fname = self.baseName + self.BILEXT
        cols = int(self.headers["NCOLS"])
        rows = int(self.headers["NROWS"])
        nbytes = int(self.headers["NBITS"]) / 8
        data = np.fromfile(fname, dtype=np.uint16)
        ndata = data.reshape((rows, cols))
        return ndata
    
    def resize (self, mat):
        return imresize(mat, 0.6)
        
    ## Test functions
    def printHeaders (self):
        for h in self.headers.items():
            print (",".join(h))
```
